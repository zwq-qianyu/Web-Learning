# matplotlib绘制柱状图和饼状图

```python
import matplotlib.pyplot as plt
```

## 柱状图--bar

```python
x = ['q1', 'q2', 'q3', 'q4']
y = [10, 30, 20, 60]
plt.bar(x, y, color='c', width=0.3)
plt.show()
```

```python
x = ['q1', 'q2', 'q3', 'q4']
y = [10, 30, 20, 60]
plt.bar(x, y, color='c', width=0.3)
plt.grid(True)
plt.show()
```

```python
x = ['q1', 'q2', 'q3', 'q4']
y = [10, 30, 20, 60]
rect = plt.bar(x, y, color='c', width=0.3)
for ind, item in enumerate(rect):
    _x = item.get_x() + 0.1
    _y = item.get_height() + 2
    plt.text(_x, _y, y[ind])
plt.ylim(0, 70)     # y轴取值范围为0到70
plt.show()
```

## 饼状图--pie

```python
x
```

```python
plt.pie(y)        # 画饼状图只需要y值
plt.show()
```

```python
plt.axes(aspect=1)        # 设置为正圆
plt.pie(y, labels=x, autopct='%.2f%%')
plt.show()
```

```python
plt.axes(aspect=1)        # 设置为正圆
plt.pie(y, labels=x, autopct='%.2f%%', explode=[0.2, 0, 0, 0])        # explode表示凸出
plt.show()
```

```python
plt.axes(aspect=1)        # 设置为正圆
plt.pie(y, labels=x, autopct='%.2f%%', explode=[0.2, 0, 0.1, 0], shadow=True)        # explode表示凸出
plt.show()
```
