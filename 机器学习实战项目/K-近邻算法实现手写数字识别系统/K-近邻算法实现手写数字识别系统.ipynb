```python
# 将图像转换为向量：该函数创建1x1024的NumPy数组，然后打开给定的文件，
# 循环读出文件的前32行，并将每行的头32个字符值存储在NumPy数组中，最后返回数组
def img2vector(filename):
    # 创建向量
    returnVect = np.zeros((1, 1024))
    
    # 打开数据文件，读取每行内容
    fr = open(filename)
    
    for i in range(32):
        # 读取每一行
        linelist = fr.readline()
        
        # 将每行前 32 字符装成 int 存入向量
        for j in range(32):
            returnVect[0, 32*i + j] = int(linelist[j])
    return returnVect
```

## 2、分析数据

#### 算法实现过程：

1、计算已知类别数据集中的点与当前点之间的距离；<br>
2、按照距离递增次序排序；<br>
3、选取与当前点距离最小的k个点；<br>
4、确定前k个点所在类别的出现频率；<br>
5、返回前k个点出现频率最高的类别作为当前点的预测分类。<br>

```python
from collections import Counter
```

```python
def distance(a, b, p):
    # 计算明可夫斯基距离
    return np.sum(np.abs(a - b) ** p) ** (1/p)
```

```python
class KNeighborsClassifier():
    # knn 分类器
    def __init__(self, k=2, p=2):
        self.k = k
        self.p = p
        
        self.x_train = None
        self.y_train = None
    
    def fit(self, x_train, y_train):
        self.x_train = x_train
        self.y_train = y_train
        
        return self
    
    def predict(self, x_test):
        return np.array([self._predict(x, self.k, self.p) for x in x_test])
    
    def _predict(self, x, k, p):
        # 预测一组数据
        distances = [distance(item, x, p) for item in self.x_train]
        nearest = np.argsort(distances)[:k]    # 取距离最小的 k 个
        #print(nearest)
        k_labels = self.y_train[nearest] 
        
        return Counter(k_labels).most_common(1)[0][0]
```

```python
def createDataSet():
    group = np.array([[1.0,1.1],[1.0,1.0],[0,0],[0,0.1]])
    labels = np.array(['A','A','B','B'])
    return group, labels
```

```python
group, labels = createDataSet()
knn = KNeighborsClassifier()
knn.fit(group, labels)
```

```python
class handWritingClassTest():
    # 训练样本数据的类标签列表
    hwlabels = []
    
    # 所有训练数据的文件名
    trainningFiles = listdir('digits/trainingDigits')
    m = len(trainningFiles)
    
    # 初始化训练样本矩阵
    trainningMatrix = np.zeros((m, 1024))
    
    for i in range(m):
        # 提取数据分类标签
        trainningFilesStr = trainningFiles[i]
        trainningFilesName = trainningFilesStr.split('.')[0]
        trainningFilesClassName = int(trainningFilesName.split('_')[0])
        hwlabels.append(trainningFilesClassName)
        
        # 提取数据向量
        trainningMatrix[i, :] = img2vector('digits/trainingDigits/%s' % trainningFilesStr)   # 必须是相对当前的完整路径
    
    
    # 测试数据的类标签列表
    test_labels = []
    
    # 所有测试数据的文件名
    testFiles = listdir('digits/testDigits')
    m_test = len(testFiles)
    
    # 初始化测试样本矩阵
    testMatrix = np.zeros((m_test, 1024))
    
    for i in range(m_test):
        # 提取文件名中的数字
        testFilesStr = testFiles[i]
        fileStr = testFilesStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        test_labels.append(classNumStr)
        
        # 提取数据向量
        testMatrix[i, :] = img2vector('digits/testDigits/%s' % testFilesStr)   # 必须是相对当前的完整路径
    
    
    # 开始训练
    hw_knn = KNeighborsClassifier(k = 3)
    hw_knn.fit(trainningMatrix, np.array(hwlabels))
    
    # 对测试数据进行预测
    predict_labels = hw_knn.predict(np.array(testMatrix))
    
    # 计算错误率
    error_num = np.sum(test_labels == predict_labels)
    error_rate = 1 - error_num / len(test_labels)
    
    print('错误个数：', error_num)
    print('错误率：', error_rate)
```
