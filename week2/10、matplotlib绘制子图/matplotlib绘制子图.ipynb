# matplotlib绘制子图

语法和MATLAB几乎相同

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2, 1, 1)    # 2 X  1
plt.subplot(2, 1, 2)
plt.show()
```

```python
plt.subplot(1, 2, 1)    # 2 X  1
plt.subplot(1, 2, 2)
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.subplot(2, 1, 2)
plt.show() 
# 先看成2*2的，然后到下面时看成2*1的
```

```python
import numpy as np

plt.subplot(2, 2, 1)
x = np.random.random(10)
y = np.random.random(10)
plt.scatter(x, y)
plt.title('test')

plt.subplot(2, 2, 2)
x = np.arange(7)
y = x**2
plt.plot(x, y)

plt.subplot(2, 1, 2)
plt.show() 
# 先看成2*2的，然后到下面时看成2*1的
```
